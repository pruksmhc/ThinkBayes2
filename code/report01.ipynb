A group of four friends are about to take a major international math competition exam similar to the International Math Olympiad but which is made up of one multiple choice question in mathematics. They pass by a room and see the Head of Examination for their school discuss the answer. Simon says he overheard the Head saying the answer is A and thinks it is the answer. The friends, since they love math, decided to do an analysis. The distribution of answers in the past are as below:  

    A: 35%
    B: 19%
    C: 20%
    D: 26%
    
Given that Simon was telling the truth (in that he indeed heard A), and human hearing at age 18 (Simonâ€™s age) is inaccurate 0.24 of the time, what is the probability that the answer is indeed A? 

```python
from __future__ import print_function, division
from thinkbayes2 import Hist, Pmf
% matplotlib inline
```

The prior is stated as above.

The original question is : given Simon heard A, what is the probaiblity that the answer is indeed A? 

Here, the set of hypothesis is binary, whether or not A is indeed the answer. H1 is the answer is indeed A, and H2 is the answer is not A. 

P(H1) = 0.35, and P(H2) = 0.65. 

The likelihood, or P(D|H1) = (0.35)(0.76), and P(D|H2) = (0.65)(0.24). 

This is because if the answer is indeed A, then Simon's hearing must have been correct (0.76 chance). If the answer is not A, then Simon's hearing was faulty (0.24 chance).

```python
pmf = Pmf()
pmf.Set('The answer to the exam is indeed A', 0.35)
pmf.Set('The answer is not A', 0.65)
pmf.Mult('The answer to the exam is indeed A', 0.76)
pmf.Mult('The answer is not A', 0.24)
pmf.Normalize()
pmf.Print()
```
